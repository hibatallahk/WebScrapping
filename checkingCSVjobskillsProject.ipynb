```python
import requests
from bs4 import BeautifulSoup

 

url ='https://www.indeed.com/jobs?q=web+developer&l=New+York'

page = requests.get(url)

 

soup = BeautifulSoup(page.content,'html.parser')
```

```python
url ='https://ma.indeed.com/jobs?q=data+scientist&l=Maroc'
page = requests.get(url)

 

soup = BeautifulSoup(page.content,'html.parser')

 

results = soup.find(id='resultsCol')
indeed_jobs = results.select('div.job_seen_beacon')
 

print(results.prettify())
```

```python
indeed_jobs
```

```python
for indeed_job in indeed_jobs:
   job_title = indeed_job.find('h2',class_='jobTitle-color-purple')

   job_company = indeed_job.find('span',class_='companyName')

   job_location = indeed_job.find('span',class_='companyLocation')
```

```python
print(job_title)
print()
print(job_company)
print()
print(job_location)
```
